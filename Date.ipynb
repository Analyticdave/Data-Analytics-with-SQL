## Using the date_trunk function

```python
%%sql

select
orderdate,
DATE_TRUNC('day', orderdate) as order_date_day
from sales
order by random()
limit 10;

```

```python
%%sql
select
orderdate,
DATE_TRUNC('month', orderdate) as order_date_day
from sales
order by random()
limit 10;
```

```python
%%sql

--cleaning up the date column
select
orderdate,
DATE_TRUNC('month', orderdate)::date as order_date_day
from sales
order by random()
limit 10;
```

```python
%%sql

-- findidng the net revenue
select
orderdate,
DATE_TRUNC('month', orderdate)::date as order_month,
sum(quantity * netprice * exchangerate) as net_revenue
from sales
group by
orderdate
limit 10;
```

```python
%%sql

-- doing a unique count of customers
select
orderdate ::date,
--DATE_TRUNC('month', orderdate)::date as order_month,
sum(quantity * netprice * exchangerate) as net_revenue,
count(distinct customerkey) as total_unique_customers
from sales
group by
orderdate
```

## using the to_char function

```python
%%sql

---Showing datestring for year using the format specifier
select
orderdate,
to_char(orderdate, 'YYYY') as order_year,
sum(quantity * netprice * exchangerate) as net_revenue,
count(distinct customerkey) as total_unique_customers
from sales
group by
orderdate
```

```python
%%sql

---Showing datestring fro year and month using the format specifier
select
orderdate,
to_char(orderdate, 'YYYY-MM') as order_year,
sum(quantity * netprice * exchangerate) as net_revenue,
count(distinct customerkey) as total_unique_customers
from sales
group by
orderdate
```

## Date and time filtering | date_part and date_now

```python
%%sql

--Filtering using Datr_part
select

orderdate,

date_part('year', orderdate) as order_year,
date_part('month', orderdate) as order_month,
date_part('day', orderdate) as order_date_day,
date_part('hour', orderdate) as order_hour,
date_part('minute', orderdate) as order_minute,
date_part('second', orderdate) as order_second

from sales

order by random()
limit 10;

```

```python
%%sql

--- Using extract function

select

extract(year from orderdate) as order_year,
extract(month from orderdate) as order_month,
extract(day from orderdate) as order_date_day,
sum(quantity * netprice * exchangerate) as net_revenue

from sales
group by order_year, order_month, order_date_day
order by order_year, order_month, order_date_day
limit 10
```

```python
%%sql

--- searching orders five years ago

select
s.orderdate,
p.categoryname,
sum(s.quantity * s.netprice * exchangerate) as net_revenue

from sales as s
left join product as p
on s.productkey = p.productkey
where
extract(year from s.orderdate) >= extract(year from CURRENT_DATE) -5
group by
s.orderdate,
p.categoryname
limit 10;
```

```python
%%sql

--- searching orders five years ago using now

select
s.orderdate,
p.categoryname,
sum(s.quantity * s.netprice * exchangerate) as net_revenue

from sales as s
left join product as p
on s.productkey = p.productkey
where
extract(year from s.orderdate) >= extract(year from now()) -5
group by
s.orderdate,
p.categoryname
limit 10;
```

```python
%%sql

select
s.orderdate,
p.categoryname,
sum(s.quantity * s.netprice * exchangerate) as net_revenue

from sales as s
left join product as p
on s.productkey = p.productkey
where
interval '5 years' >=  now() - s.orderdate
group by
s.orderdate,
p.categoryname
limit 10;
```
